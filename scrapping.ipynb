```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from pprint import pprint
from lxml import html
from lxml.html import fromstring
import urllib.request
from urllib.request import urlopen
import random
import re
import scrapy
```

```python
def webScraper(url):
    res = requests.get(url)
    html = res.text
    return BeautifulSoup(html, 'html.parser')
```

```python
url = 'https://www.rottentomatoes.com/top/bestofrt/top_100_drama_movies/'
```

```python
soup = webScraper(url)
```

```python
movieslist = []
moviesraw = soup.find_all('a',{"class": "unstyled articleLink"})
for movie in moviesraw:
    movie = movie.text
    movie = re.findall(r'\n.+\)', movie)
    movieslist.append(movie)
movieslist
```

```python
prefinalist = [item for sublist in movieslist for item in sublist]
comedymovies = []
for movie in prefinalist:
    movie = re.sub("\n", "", movie)
    movie = re.sub("\(\d+\)", "", movie)
    movie = movie.strip()
    comedymovies.append(movie)
print(len(comedymovies))
    

    
```
